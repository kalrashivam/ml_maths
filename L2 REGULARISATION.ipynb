```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
N = 50
X = np.linspace(0,10,N)
Y = 0.5*X + np.random.randn(N)
```

```python
Y[-1] +=30
Y[-2] +=35
```

```python
plt.scatter(X,Y)
plt.show()
```

```python
X = np.vstack([np.ones(N), X]).T
```

```python
w_ml = np.linalg.solve(X.T.dot(X), X.T.dot(Y))
Yans = X.dot(w_ml)
```

```python
plt.scatter(X[:,1], Y)
plt.plot(X[:,1], Yans)
plt.show()
```

```python
l2 = 1000.0
w_map = np.linalg.solve(l2*np.eye(2) + X.T.dot(X), X.T.dot(Y))
Yhat_map = X.dot(w_map)
```

```python
plt.scatter(X[:,1], Y)
plt.plot(X[:,1], Yans, label='maximum likelihood')
plt.plot(X[:,1], Yhat_map, label='map')
plt.legend()
plt.show()
```
